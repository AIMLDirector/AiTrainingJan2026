```python
from nltk.tokenize import word_tokenize
sentence = "Garak is an open-source, audit-style LLM vulnerability scanner developed with support from NVIDIA. It is compared to network scanners like nmap in its function. It applies a wide range of pre-defined probes and attack types based on academic research and documented weaknesses to find known flaws"
sentence = sentence.lower()
print(sentence)
```

```python
word_sentence_cleaned = word_tokenize(sentence)
print(word_sentence_cleaned)
```

We are going to do the cleanup of the punctuation and converting into lower


```python
word_sentence_lower = [word.lower() for word in word_sentence_cleaned]
print(word_sentence_lower)
```

```python
import string
word_sentence_punct = [word for word in word_sentence_lower if word not in string.punctuation]
print(word_sentence_punct)
```

```python
word_cleaned = []
for i in word_sentence_lower:
    if i not in string.punctuation:
        word_cleaned.append(i)
print(word_cleaned)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
word_sentence_stopword = [word for word in word_sentence_punct if word not in stop_words]
print(word_sentence_stopword)
```

```python
from nltk.stem import SnowballStemmer,PorterStemmer, LancasterStemmer
stemmer = PorterStemmer()

porter_word = [stemmer.stem(word) for word in word_sentence_stopword]
print(porter_word)
```

```python
lancer = LancasterStemmer()
lance_word = [lancer.stem(word) for word in word_sentence_stopword]
print(lance_word)
```

```python
snowball =SnowballStemmer(language='english')
snowball_word = [snowball.stem(word) for word in word_sentence_stopword]
print(snowball_word)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer = [lemmatizer.lemmatize(word) for word in word_sentence_stopword]
print(lemmatizer)
```

```python
from nltk.stem import SnowballStemmer, PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

snowball = SnowballStemmer(language='english')
porter = PorterStemmer()
wordnet = WordNetLemmatizer()

text = ["better", "Caring", "are", "am", "worse", "struggling", 'meeting']

print("{:<12}{:<20}{:<20}{:<20}".format("Word", "Snowball Stemmer", "Porter Stemmer", "Wordnet Lemmatizer"))

for token, tag in pos_tag(text):
    pos = tag[0].lower()
    if pos not in ['a', 'r', 'n', 'v']:
        pos = 'n'
    print("{:<12}{:<20}{:<20}{:<20}".format(
        token,
        snowball.stem(token),
        porter.stem(token),
        wordnet.lemmatize(token, pos)
    ))
```
